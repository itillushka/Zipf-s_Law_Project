```python
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
def read_book(file_path):
    with open(file_path, 'r') as file:
        text = file.read()
    words = text.split()
    return Counter(words)
```

```python
def clear_punctuation(text):
    punctuation = [".", ",", "!", "?", ";", ":", "_", "(", ")", "[", "]", "{", "}", "<", ">", "/", "\\", "|", "*", "&", "^", "%", "$", "#", "@", "~", "`", "\"", "'","‚Äù"]
    for p in punctuation:
        text = text.replace(p, " ")
    return text
```

```python
def count_words(text):
    words = text.split()
    word_count = {}
    for word in words:
        word_count[word] = word_count.get(word, 0) + 1
    return sorted(word_count.items(), key=lambda x: x[1], reverse=True)
```

```python
def calculate_frequencies(word_count):
    total = sum([count for word, count in word_count])
    frequencies = {}
    for word, count in word_count:
        frequencies[word] = count / total
    return frequencies
```

```python

def plot_word_frequencies(text):
    word_count = count_words(text)
    frequencies = calculate_frequencies(word_count)
    
    sorted_words = sorted(frequencies.items(), key=lambda item: item[1], reverse=True)
    ranks = np.arange(1, len(word_count) + 1)
    freqs = [freq for word, freq in sorted_words]

    plt.figure(figsize=(10, 6))
    plt.plot(ranks, freqs, marker='o',markersize=1, linestyle='-')
    plt.xscale('log')
    plt.yscale('log')
    plt.ylim(0.000005, 0.1)  
    plt.xlim(1, 50000)    
    plt.xlabel('Rank')
    plt.ylabel('Frequency')
    plt.title('Word Frequencies in Log-Log Scale')
    plt.grid(True, which="both", ls="--")
    plt.show()

content = read_book("Book1_Wuthering_Heights.txt")
clear_content = clear_punctuation(content)
plot_word_frequencies(clear_content)
```

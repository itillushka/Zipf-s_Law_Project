```python
def read_book(file_path):
    file = open(file_path, "r", encoding="utf8")
    content = file.read()
    file.close()
    return content
```

```python
def clear_punctuation(text):
    punctuation = [".", ",", "!", "?", ";", ":", "_", "(", ")", "[", "]", "{", "}", "<", ">", "/", "\\", "|", "*", "&", "^", "%", "$", "#", "@", "~", "`", "\"", "'","”"]
    for p in punctuation:
        text = text.replace(p, " ")
    return text

content = read_book("Book1_Wuthering_Heights.txt")
clear_content = clear_punctuation(content)
```

```python
def count_words(text):
    words = text.split()
    word_count = {}
    for word in words:
        word_count[word] = word_count.get(word, 0) + 1
    return sorted(word_count.items(), key=lambda x: x[1], reverse=True)

word_count = count_words(clear_content)
print(word_count)
```
